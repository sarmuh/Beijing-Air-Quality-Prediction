```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge

# 1. Ma'lumotlarni o'qish
df = df.copy()

# 2. LAG FEATURES YARATISH (Barcha kerakli ustunlar uchun)
# DIQQAT: WSPM, TEMP va DEWP ham qo'shildi!
lag_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'WSPM', 'TEMP', 'DEWP']
for col in lag_columns:
    df[f'{col}_shift24'] = df.groupby('station')[col].shift(24)

# Kelajakni ko'rib qolmaslik (Data Leakage) uchun asl gaz ustunlarini o'chiramiz
cols_to_drop = ['PM10', 'SO2', 'NO2', 'CO', 'O3']
df = df.drop(columns=cols_to_drop)

# 3. FUTURE ENGINEERING (Barcha qatorlar joyida turgan holatda yasaladi)
df['PM2.5_rolling_mean_24'] = df.groupby('station')['PM2.5_shift24'].transform(lambda x: x.rolling(window=24, min_periods=1).mean())
df['WSPM_rolling_mean_24'] = df.groupby('station')['WSPM_shift24'].transform(lambda x: x.rolling(window=24, min_periods=1).mean())
df['PM2.5_trend'] = df['PM2.5_shift24'] - df.groupby('station')['PM2.5'].shift(48)
df['Temp_Dewp_Diff'] = df['TEMP_shift24'] - df['DEWP_shift24']

# 4. ENG OXIRIDA XATOLARNI TOZALASH
# Trend uchun 48 soatlik tarix ishlatilgani sababli, birinchi 48 soatni kesamiz
df = df[df.groupby('station').cumcount() >= 48].reset_index(drop=True)

# 5. TRAIN VA TEST GA AJRATISH (Vaqt bo'yicha)
train_data = df[df['year'] < 2017].copy()
test_data = df[df['year'] == 2017].copy()

# 6. ITERATIVE IMPUTER BILAN XAVFSIZ TO'LDIRISH
# Faqat raqamli ustunlarni filtrlaymiz
numeric_cols = train_data.select_dtypes(include=['float64', 'int64']).columns

imputer = IterativeImputer(
    estimator=BayesianRidge(verbose=1),          
    max_iter=20,             
    random_state=42,        
)

# Train'da qonuniyatni o'rganib, Train'ni to'ldiramiz
train_numeric_filled = imputer.fit_transform(train_data[numeric_cols])

# O'sha o'rganilgan bilim bilan Test'ni to'ldiramiz (fit qilmasdan!)
test_numeric_filled = imputer.transform(test_data[numeric_cols])

# To'ldirilgan raqamlarni asl jadvallarga qaytarib joylaymiz
train_cleaned = train_data.copy()
test_cleaned = test_data.copy()

train_cleaned[numeric_cols] = train_numeric_filled
test_cleaned[numeric_cols] = test_numeric_filled

print("Tabriklayman! Ma'lumotlar birorta ham NaN qoldirmasdan, eng xavfsiz va aqlli yo'l bilan ML model uchun tayyorlandi.")
```
